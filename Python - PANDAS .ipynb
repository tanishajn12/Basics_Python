```python
import pandas as pd 
  
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']} 

data2 = {'Name':['Gaurav', 'Anuj', 'Dhiraj', 'Hitesh'], 
        'Age':[22, 32, 12, 52], 
        'Address':['Allahabad', 'Kannuaj', 'Allahabad', 'Kannuaj'], 
        'Qualification':['MCA', 'Phd', 'Bcom', 'B.hons'],
        'Salary':[1000, 2000, 3000, 4000]} 

df1 = pd.DataFrame(data1,index=[0, 1, 2, 3])
 
df2 = pd.DataFrame(data2, index=[2,3,6,7])

ds=pd.concat([df1,df2],axis=1, join='inner')
ds
```

```python
df3=df1.append(df2)
df3
```

```python
df4 = pd.concat([df1, df2], ignore_index=True)
df4
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
         'key1': ['K0', 'K1', 'K0', 'K1'],
        'Age':[27, 24, 22, 32],} 
   

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
          'key1': ['K0', 'K0', 'K0', 'K0'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2) 
res=pd.merge(df1,df2,on='key')
res
```

```python
res1=pd.merge(df1,df2,on=['key','key1'])
res1
```

left	LEFT OUTER JOIN-----Use keys from left frame only
right	RIGHT OUTER JOIN-----Use keys from right frame only
outer	FULL OUTER JOIN-----Use union of keys from both frames
inner	INNER JOIN-----Use intersection of keys from both frames

```python
rdf=pd.merge(df1,df2,how='left',on=['key','key1'])
rdf
```

```python
df=pd.merge(df1,df2,how='right',on='key')
df
```

```python
res3=pd.merge(df1,df2,how='outer',on=['key','key1'])
res3
```

```python
res3.dropna()
```

```python
res3.drop_duplicates('key')
```

```python
res4=pd.merge(df1,df2,how='inner',on=['key','key1'])
res4
```

```python
import pandas as pd

data = {
    'City': ['New York', 'New York', 'Los Angeles', 'Los Angeles', 'Chicago', 'Chicago'],
    'Year': [2020, 2021, 2020, 2021, 2020, 2021],
    'Population': [8.4, 8.5, 4.0, 4.1, 2.7, 2.8],
    'GDP': [1.5, 1.6, 1.0, 1.2, 0.8, 0.9]}
df = pd.DataFrame(data)
df
```

```python
df2=df.groupby('City')
agg_df=df2.agg({'Population':'mean','GDP':'sum','Year':'count'})
agg_df
```

```python
df3=df.agg({'Population':'median','GDP':'max','Year':'count'})
df3
```

mean(): Compute the mean of each column or row.
median(): Compute the median of each column or row.
min(): Find the minimum value of each column or row.
max(): Find the maximum value of each column or row.
std(): Compute the standard deviation of each column or row.
var(): Compute the variance of each column or row.
count(): Count non-null values in each column or row.
sum(): Compute the sum of each column or row.
